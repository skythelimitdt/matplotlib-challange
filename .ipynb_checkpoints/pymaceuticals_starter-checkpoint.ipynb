### Analysis

- The mean and median values for each drug regimen indicate slight skewness in the data. For some drug regimens, the mean and median are closely aligned, suggesting a more symmetric distribution, while others—such as Ketapril, Naftisol, Placebo, Propriva, Stelassyn, and Zoniferol—display greater skewness. The standard deviation, which shows the average distance of each data point from the mean, reveals that Ramicane has the smallest spread, closely followed by Capomulin. Additionally, the small standard error for these two regimens suggests that their mean values are reliable estimates of the population mean in our dataset.
- Among the four selected drug regimens (Capomulin, Ramicane, Infubinol, and Ceftamin), Infubinol displayed an outlier, indicating an extreme value in the dataset. This warrants further investigation to determine whether it should be included in statistical testing.
- A correlation coefficient of 0.84 indicates a strong positive correlation between mouse weight and average tumor volume. This suggests that as mouse weight increases, the average tumor volume also tends to increase.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem
import seaborn as sns
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data=pd.merge(study_results, mouse_metadata,how="left", on="Mouse ID")

# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
mouse_count=mouse_metadata["Mouse ID"].count()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicates = combined_data.loc[combined_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
unique_mouse_ids=duplicates["Mouse ID"].unique()
result=np.array(unique_mouse_ids)
result
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mouseid = combined_data.loc[combined_data["Mouse ID"] == "g989", :]
duplicate_mouseid
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_data=combined_data.loc[combined_data["Mouse ID"] != "g989", :]
clean_combined_data
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count_cleaned=clean_combined_data["Mouse ID"].nunique()
mouse_count_cleaned
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

mean=(clean_combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"]).mean()
median=(clean_combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"]).median()
var=(clean_combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"]).var()
std_dev=(clean_combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"]).std()
sem = clean_combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

tumor_volume_results = pd.DataFrame({
    "Mean Tumor Volume": mean,
    "Median Tumor Volume": median,
    "Tumor Volume Var": var,
    "Tumor Volume Std Dev": std_dev,
    "SEM": sem
})
                         
tumor_volume_results
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)                     

# Using the aggregation method, produce the same summary statistics in a single line
tumor_volume_per_regimen2= clean_combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(
    mean=np.mean,
    median=np.median,
    var=np.var,
    std=np.std,
    sem=lambda x: np.std(x) / np.sqrt(len(x))
).reset_index()


tumor_volume_per_regimen2    
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

filtered_drug_regimen=clean_combined_data.loc[(clean_combined_data["Drug Regimen"] == "Capomulin" ) | 
                                        (clean_combined_data["Drug Regimen"] == "Ramicane" ) |
                                        (clean_combined_data["Drug Regimen"] == "Infubinol" ) |
                                        (clean_combined_data["Drug Regimen"] == "Ceftamin" ), :]

# Start by getting the last (greatest) timepoint for each mouse
greatest_time_point=filtered_drug_regimen.groupby(["Drug Regimen", "Mouse ID"])["Timepoint"].max()
greatest_time_point_permouse=pd.DataFrame(greatest_time_point)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data=pd.merge(greatest_time_point_permouse, filtered_drug_regimen, on=["Drug Regimen", "Mouse ID", "Timepoint"], how="left")

merged_data


```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = merged_data["Drug Regimen"].unique().tolist()

# Create empty list to fill with tumor vol data (for plotting)
treatment_labels = []
tumor_vol_data = []
mouse_ids = []

# Step 3: Loop through each treatment and collect the data
for treatment in treatments:
    treatment_data = merged_data[merged_data["Drug Regimen"] == treatment]
    treatment_labels.extend([treatment] * len(treatment_data))
    tumor_vol_data.extend(treatment_data["Tumor Volume (mm3)"].tolist())
    mouse_ids.extend(treatment_data["Mouse ID"].tolist())

# Step 4: Create a DataFrame with the collected data
tumor_df = pd.DataFrame({
    'Drug Regimen': treatment_labels,
    'Final Tumor Volume (mm3)': tumor_vol_data,
    'Mouse ID': mouse_ids
})

# Step 5: Identify outliers for each treatment
outliers_dict = {}  # Dictionary to store outliers for each treatment

for treatment in treatments:
    # Filter data for the current treatment
    treatment_data = tumor_df[tumor_df["Drug Regimen"] == treatment]
    
    # Calculate IQR
    Q1 = treatment_data["Final Tumor Volume (mm3)"].quantile(0.25)
    Q3 = treatment_data["Final Tumor Volume (mm3)"].quantile(0.75)
    IQR = Q3 - Q1
    # Determine bounds for outliers
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
    # Identify outliers
    outliers = treatment_data[(treatment_data["Final Tumor Volume (mm3)"] < lower_bound) | 
                              (treatment_data["Final Tumor Volume (mm3)"] > upper_bound)]
    
    # Store outliers in the dictionary
    outliers_dict[treatment] = outliers

    # Print outliers if any
    if not outliers.empty:
        print(f"Outliers for {treatment}:")
        print(outliers['Final Tumor Volume (mm3)'].to_string(index=False))
    else:
        print(f"No outliers found for {treatment}.")

```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
plt.figure(figsize=(6, 6))
# Customize the appearance of outliers (fliers)
flierprops = dict(marker='o', markerfacecolor='red', markersize=6, linestyle='none')
sns.boxplot(x='Drug Regimen', y='Final Tumor Volume (mm3)', data=tumor_df, flierprops=flierprops)
plt.ylabel('Final Tumor Volume (mm3)', fontsize=12)

# Show the plot
plt.show()
```

```python
# Create the dataframe for mouse weight vs the avg observed tumor volume
capomulin_mouseweight_df=clean_combined_data.loc[(clean_combined_data["Drug Regimen"] == "Capomulin" )]
avg_tumor_volume=capomulin_mouseweight_df.groupby("Mouse ID")["Tumor Volume (mm3)"].mean().reset_index()
avg_tumor_volume.columns = ['Mouse ID', 'Tumor Volume (mm3)']
mouse_data = capomulin_mouseweight_df[['Mouse ID', 'Weight (g)']].drop_duplicates()
data_for_plot = pd.merge(mouse_data, avg_tumor_volume, on='Mouse ID')
data_for_plot
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
plt.figure(figsize=(8, 6))
plt.scatter(data_for_plot['Weight (g)'], data_for_plot['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
mouse_weight=data_for_plot["Weight (g)"]
average_tumor_volume=data_for_plot["Tumor Volume (mm3)"]
print(f"The correlation coefficient between mouse weight and average tumor volume is {round(st.pearsonr(mouse_weight,average_tumor_volume)[0],2)}")
(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight, average_tumor_volume)
regress_values = mouse_weight * slope + intercept
plt.scatter(mouse_weight,average_tumor_volume)
plt.plot(mouse_weight,regress_values,"r-")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
